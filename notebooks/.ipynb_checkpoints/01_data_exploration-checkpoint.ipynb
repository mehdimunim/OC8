# Notebook 1 : Exploration des données


L'objectif de ce projet est de réaliser la segmentation d'images issues de caméras embarquées (voitures).

Dans ce notebook, nous explorons les données Cityscape que nous avons préalablement téléchargées :

- **Données brutes** : https://s3-eu-west-1.amazonaws.com/static.oc-static.com/prod/courses/files/AI+Engineer/Project+8+-+Participez+%C3%A0+la+conception+d'une+voiture+autonome/P8_Cityscapes_gtFine_trainvaltest.zip

- **Images segmentées/annotées** : https://s3-eu-west-1.amazonaws.com/static.oc-static.com/prod/courses/files/AI+Engineer/Project+8+-+Participez+%C3%A0+la+conception+d'une+voiture+autonome/P8_Cityscapes_leftImg8bit_trainvaltest.zip

- **Aperçu du dataset** : https://www.cityscapes-dataset.com/dataset-overview/


## Importation des bibliothèques

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

##  Chargement des données

```python
# Chemins vers les dossiers principaux
leftImg8bit_dir = "../data/raw/leftImg8bit"
gtFine_dir = "../data/raw/gtFine"

# Afficher les sous-dossiers
print("Sous-dossiers dans leftImg8bit :")
print(os.listdir(leftImg8bit_dir))

print("Sous-dossiers dans gtFine :")
print(os.listdir(gtFine_dir))
```

```python
# Exemple : Afficher les fichiers d'une ville dans train
city = "cologne"
image_files = os.listdir(os.path.join(leftImg8bit_dir, "train", city))
gt_files = os.listdir(os.path.join(gtFine_dir, "train", city))

print(f"Fichiers d'images pour {city} :")
print(image_files[:5])

print(f"Fichiers d'annotations pour {city} :")
print(gt_files[:5])
```

## Visualisation des images et masques

```python
# Chemins vers un exemple d'image et ses masques
image_path = os.path.join(leftImg8bit_dir, "train", city, "cologne_000000_000019_leftImg8bit.png")
label_ids_path = os.path.join(gtFine_dir, "train", city, "cologne_000000_000019_gtFine_labelIds.png")
color_path = os.path.join(gtFine_dir, "train", city, "cologne_000000_000019_gtFine_color.png")

# Charger les fichiers
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convertir de BGR à RGB

label_ids = cv2.imread(label_ids_path, cv2.IMREAD_UNCHANGED)
color_mask = cv2.imread(color_path)
color_mask = cv2.cvtColor(color_mask, cv2.COLOR_BGR2RGB)

# Afficher les fichiers
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title("Image originale")

plt.subplot(1, 3, 2)
plt.imshow(label_ids, cmap='tab20')
plt.title("Masque (Label IDs)")

plt.subplot(1, 3, 3)
plt.imshow(color_mask)
plt.title("Masque coloré")

plt.show()
```

## Caractéristique d'une image 

```python
# Chemins vers un exemple d'image et ses masques
image_path = os.path.join(leftImg8bit_dir, "train", city, "cologne_000000_000019_leftImg8bit.png")

# Charger les fichiers
image = cv2.imread(image_path)

# Vérifier si l'image a été chargée correctement
if image is None:
    print(f"Erreur: Impossible de charger l'image depuis {image_path}")
else:
    # Afficher les informations sur l'image
    print(f"Informations sur l'image : {image_path}")
    print(f"Dimensions : {image.shape}")
    print(f"Type de données : {image.dtype}")
    print(f"Valeur pixel maximale : {np.max(image)}")
    print(f"Valeur pixel minimale : {np.min(image)}")
    print(f"Valeur pixel moyenne : {np.mean(image)}")
    print(f"Valeur pixel médiane : {np.median(image)}")
    print(f"Écart-type des pixels : {np.std(image)}")

    # Afficher un résumé des statistiques de l'image
    print("\nRésumé des statistiques de l'image :")
    print(f"  - Moyenne des pixels : {np.mean(image)}")
    print(f"  - Médiane des pixels : {np.median(image)}")
    print(f"  - Écart-type des pixels : {np.std(image)}")
    print(f"  - Minimum des pixels : {np.min(image)}")
    print(f"  - Maximum des pixels : {np.max(image)}")
```

## Analyse des classes présentes

```python
# Identifier les classes uniques dans le masque
unique_classes = np.unique(label_ids)
print("Classes uniques dans le masque :", unique_classes)
```

Plusieurs classes présentes dans les images annotées.
Nous nous contenterons ultimement de 8 classes

## Distribution des classes

```python
# Compter le nombre de pixels par classe
class_counts = {cls: np.sum(label_ids == cls) for cls in unique_classes}
print("Distribution des pixels par classe :", class_counts)

# Affichage sous forme de graphique
plt.bar(class_counts.keys(), class_counts.values())
plt.xlabel("Classe")
plt.ylabel("Nombre de pixels")
plt.title("Distribution des classes dans le masque")
plt.show()
```

## Exploration globale

```python
# Parcourir toutes les images d'une ville et afficher quelques exemples
for i, image_file in enumerate(image_files[:3]):  # Limiter à 3 exemples pour ne pas surcharger l'affichage
    image_path = os.path.join(leftImg8bit_dir, "train", city, image_file)
    label_ids_path = os.path.join(gtFine_dir, "train", city, image_file.replace("_leftImg8bit.png", "_gtFine_labelIds.png"))

    # Charger les fichiers
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    label_ids = cv2.imread(label_ids_path, cv2.IMREAD_UNCHANGED)

    # Afficher l'image et le masque associé
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(image)
    plt.title(f"Image {i+1}")

    plt.subplot(1, 2, 2)
    plt.imshow(label_ids, cmap='tab20')
    plt.title(f"Masque {i+1}")

    plt.show()
```
